```python
import numpy as np
import pandas as pd
import difflib #difflib = if user gives spelling mistakes, it matchees with the closest match in the dataset
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
data = pd.read_csv('/content/movies.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
# prompt: return all the rows where atleast one column has a null value

null_rows = data[data.isnull().any(axis=1)]
print(null_rows)
```

```python
selected_features = ['genres', 'keywords', 'original_language', 'overview', 'original_title', 'tagline', 'director']
```

```python
for feature in selected_features:
  data[feature] = data[feature].fillna(' ')
```

```python
combined_features = data['genres']+' '+data['keywords']+' '+data['original_language']+' '+data['overview']+' '+data['original_title']+' '+data['tagline']+' '+data['director']
print(combined_features)
```

```python
vec = TfidfVectorizer()
```

```python
combined_vector = vec.fit_transform(combined_features)
print(combined_vector)
```

```python
similarity_score = cosine_similarity(combined_vector)
print(similarity_score)
```

```python
print(similarity_score.shape)
```

```python
Chosen_movie = input('Enter your favourite movie: ')
```

```python
corrected_list = difflib.get_close_matches(Chosen_movie, data['original_title'])
print(corrected_list)
```

```python
corrected_movie = corrected_list[0]
print(corrected_movie)
```

```python
index_of_movie = data[data.original_title == corrected_movie]['index'].values[0]
print(index_of_movie)
```

```python
similar_movies_score = list(enumerate(similarity_score[index_of_movie]))
print(similar_movies_score)
```

```python
sorted_order = sorted(similar_movies_score, key = lambda x:x[1], reverse = True)
print(sorted_order)
```

```python
print('Suggested movies for you: ')
i=1
for movie in sorted_order:
  index = movie[0]
  title_from_index = data[data.index==index]['original_title'].values[0]
  if i<30:
    print(i,'.', title_from_index)
    i=i+1
```
